```python
import cv2
import numpy as np
import csv
import os
import json 
# Load the CSV file
csv_file = 'mainSelected.csv'


image_path = '/home/ishant/Desktop/Drone/frames2/frame_0000.jpg'
image = cv2.imread(image_path)

main_x = image.shape[1]
main_y = image.shape[0]

frame_x = [0, main_x, main_x, 0]  # Define x coordinates for a rectangle covering the whole width
frame_y = [0, 0, main_y, main_y]
polygon_points_dict = {}

# Read CSV and store polygon points
with open(csv_file, mode='r') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        frame_name = row['filename']
        # print(row['region_shape_attributes'])
        polygon_dict = json.loads(row['region_shape_attributes'])
        # print(polygon_dict)
        if 'all_points_x' in polygon_dict and 'all_points_y' in polygon_dict:
           
            polygon_points_x = list(map(int, polygon_dict['all_points_x']))
            polygon_points_y = list(map(int, polygon_dict['all_points_y']))
            polygon_points = list(zip(polygon_points_x, polygon_points_y))
        else:
           
            polygon_points_x = frame_x
            polygon_points_y = frame_y
            polygon_points = list(zip(polygon_points_x, polygon_points_y))

        polygon_points_dict[frame_name]=polygon_points
# print(polygon_points_dict)
```

```python

frame_files = sorted([f for f in os.listdir('frames2') if f.endswith('.jpg')])
prev_mask = []
for frame_file in frame_files:
    # print(frame_file)
    frame_path = os.path.join('frames2', frame_file)

  
    image = cv2.imread(frame_path)
    mask = np.zeros_like(image)

    
    polygon_points = polygon_points_dict.get(frame_file, [])
    # Draw the polygon on the mask
    if polygon_points:
        cv2.fillPoly(mask, [np.array(polygon_points, dtype=np.int32)], (255, 255, 255))
    else:
        mask = prev_mask

  
    mask_filename = os.path.join('mask', frame_file.replace(".jpg", ".jpg"))
    cv2.imwrite(mask_filename, mask)
    prev_mask = mask
```
